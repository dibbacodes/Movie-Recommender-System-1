```python
def convert (obj):
    L = []
    for i in obj:
        L.append(i['name'])
    return L
```

```python
convert(movies.iloc[0].genres)
```

```python
import ast
#using literal eval funtion to convert string to list
ast.literal_eval(movies.iloc[0].genres)
```

```python
def convert (obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#extracting top 3 actor names
def convert2 (obj):
    L = []
    cnt=0;
    for i in ast.literal_eval(obj):
        if(cnt==3):
            break;
        else:
            L.append(i['name'])
            cnt+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break;
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# overview is still a string we want to convert it to a list as well
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
# We now want to remove spaces between names to avoid confusion in 
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']  
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df.head()
```

```python
# converting the list: tags to string
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head() 
```

Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
# do steming for similar words
#[ love, loved, loving] = [love, love, love]

import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
#examples of stemming
ps.stem('dancing')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags']
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
 #cosine distance
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#4806x4806 matrix with each row suggesting the similarity of a movie with others
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
# we want to pair the similarity with index then sort
list(enumerate(similarity[0])) # this tells 0 index movie ka har movie index sath similarity
```

```python
#sort on basis of similarity
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])
```

```python
def recommend (movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Tangled')
```
